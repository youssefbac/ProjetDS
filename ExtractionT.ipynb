```python
bounds = reader.readtext(np.array(images[0]), min_size=0, slope_ths=0.2, ycenter_ths=0.7, height_ths=0.6, width_ths=0.8,decoder='beamsearch', beamWidth=10)
bounds
```

```python
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(images[0], bounds)
```

###Processing Text

# Tokenization
Splitting a text into meaningful segments called toknizer
the input is unicode text and the output is a Doc Object

```python
text=''
for i in range(len(bounds)):
  text = text + bounds[i][1] +'\n'

print(text)
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
for token in doc:
  print(token.text)
```

####Lemmatazation

Assigning the base form of the world

```python
for token in doc:
  print(f'{token.text:{15}} {token.lemma_:{15}} {token.pos_:{15}} {token.is_stop}') #token.is_stop return true end of line
```

####Syntactic Dependecy Parsing

Chunking is extracting the important phrases present in our text. 

```python
for chunk in doc.noun_chunks:
  print(f'{chunk.text:{30}} {chunk.root.text:{30}} {chunk.root.dep_} ')
```
